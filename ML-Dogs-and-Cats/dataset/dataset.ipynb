Após o download da biblioteca, execute o comando abaixo para buscar e usar a chave do kaggle no arquivo *.env*

```python
import os
from dotenv import load_dotenv

caminho_env = '../.env'
load_dotenv(caminho_env)

api_key_kaggle = os.getenv('KAGGLE_API_TOKEN')

if not api_key_kaggle:
    raise ValueError("A chave da API não foi encontrada! Verifique se o arquivo .env está dentro da pasta 'ML-Dogs-and-Cats'")
```

Execute o comando abaixo para realizar o download do dataset que iremos utilizar.

Vale lembrar que estamos construíndo um dataset com imagens de cachorros e de gatos para o treinamento da nossa ML

```python
import kaggle

path_dataset='.'

kaggle.api.authenticate()

print(kaggle.api.dataset_list_files('anthonytherrien/dog-vs-cat').files)

kaggle.api.dataset_download_files('anthonytherrien/dog-vs-cat', path=path_dataset, unzip=True)

datasets = kaggle.api.dataset_list(search='anthonytherrien/dog-vs-cat')
print(datasets)
```

Com o download já realizado, iremos migrar as subpastas *cat* e *dog* para dentro da pasta *dataset*

```python
import shutil
from pathlib import Path

pasta_origem = Path('./animals')
pasta_destino = Path('.')

try:
    for subpasta in pasta_origem.iterdir():
        if subpasta.is_dir():
            shutil.move(str(subpasta), str(pasta_destino))
            print(f"Subpasta {subpasta.name} movida para {pasta_destino}")
except Exception as e:
    print(f'Erro na execução.')
```

Apagar a pasta *animals*, já que ela está vazia caso não houver problemas com a execução da célula anterior

```python
import os

pasta_vazia = './animals'

try:
    os.rmdir(pasta_vazia)
    print(f'A pasta no caminho \'{pasta_vazia}\' foi apagada com sucesso!')
except OSError as e:
    print(f'Erro: {e.strerror} \t-\t{e.filename}')
```

Renomear as pastas **cat** e **dog** para **Gato** e **Cachorro**, respectivamente

```python
import os
from pathlib import Path

dataset = Path('.')
novo_nome_subpasta = ['Gato', 'Cachorro']
cont = 0

try:
    for subpasta_antiga in dataset.iterdir():
        if subpasta_antiga.is_dir():
            os.rename(subpasta_antiga, novo_nome_subpasta[cont])
            print(f"Pasta '{subpasta_antiga}' renomeada para '{novo_nome_subpasta[cont]}' com sucesso.")
            cont += 1

except FileNotFoundError:
    print("A pasta original não foi encontrada.")
except OSError as e:
    print(f"Erro ao renomear: {e}")
```
